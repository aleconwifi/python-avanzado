### Importación de librerías

```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('ggplot')
import numpy as np
```

### Importación del dataset

```python
df = pd.read_csv('accidents_2017.csv')
df.head()
```



### Limpieza de Datos

```python
df.columns
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.replace('Unknown',np.nan, inplace=True)
```

```python
df.info()
```

```python
df.drop(['District Name','Neighborhood Name', 'Street'],axis=1 ,inplace=True)
```

reemplazar los nombres de las columnas: los espacios por guiones bajos y mayúsculas por minúsculas. Además, incluimos una columna con el año.

```python
df.rename(columns=lambda x:x.replace(' ','_').lower(), inplace=True)
df['year'] = np.repeat(2017,df.shape[0])
df.head()
```

```python
df.info()
```

Como podemos observar, Mes, Día y Hora no son objetos de fecha y hora. Podemos combinar fácilmente esas columnas en una sola columna usando la función pandas.to_datetime. Ahora, podemos combinar día de la semana, mes, día, hora y año en una sola columna llamada fecha. Para evitar un ValueError, tenemos que convertir los nombres de los meses en enteros antes de usar la función pandas.to_datetime.

```python
list(df.month.unique())
```

```python
df['month_str'] = df['month']
month_to_int = {'January':1,'February':2,'March':3,'April':4,'May':5,'June':6,'July':7,'August':8,'September':9,'October':10,'November':11,'December':12}
df['month'].replace(month_to_int,inplace=True)
list(df.month.unique())
```

```python
df.head()
```

```python
df['date']=pd.to_datetime(df[['year', 'month', 'day','hour']])
df.head()
```

Después de la conversión, podemos obtener una columna de fecha y hora utilizando la función mencionada de la siguiente manera:

```python
df['date'].dt.dayofweek
```

Para acceder fácilmente a la información sobre un accidente automovilístico, vamos a establecer la identificación (Id) como el índice del marco de datos, eliminando antes de los espacios finales presentes en las entradas de identificación.


```python
df.id.loc[0]
```

```python
df.id = df.id.apply(lambda x: x.strip())
df.set_index('id', inplace=True)

df.loc['2017S008429']
```

```python
df
```

El paso final de limpieza consiste en evaluar si hay entradas duplicadas en el marco de datos. Si es así, eliminaremos estas entradas duplicadas del marco de datos, ya que representan el mismo accidente automovilístico.

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

### Gráficos con pandas

### Cantidad de accidentes automovilísticos por mes

Para analizar la distribución de accidentes automovilísticos por mes, empleamos la función pandas.DataFrame.groupby.


```python
accidentes_mes = df.groupby(df['month_str']).count().date
accidentes_mes
```

```python
import calendar
accidentes_mes = df.groupby(df['month']).count().date
accidentes_mes.index=[calendar.month_name[x] for x in range(1,13)]
accidentes_mes = accidentes_mes.sort_values()
accidentes_mes
```

```python
accidentes_mes.plot(kind='bar',figsize=(12,7), color='red', alpha=0.5)
plt.title('Accidentes en Barcelona en 2017', fontsize=20)
plt.xlabel('Mes',fontsize=16)
plt.ylabel('Número de Accidentes',fontsize=16);
```

```python
plt.figure(figsize=(12,7))
plt.bar(accidentes_mes.index, accidentes_mes, color='r', alpha=0.5) 
plt.title('Accidentes en Barcelona en 2017', fontsize=20)
plt.xlabel('Mes',fontsize=16)
plt.ylabel('Número de Accidentes',fontsize=16);
```

Como podemos observar, el número de accidentes disminuye en agosto y diciembre. Una razón podría ser que menos personas conducen al trabajo en estos meses.

### Cantidad de accidentes automovilísticos por día de la semana

Como lo hicimos con los meses, también podemos analizar la distribución de los accidentes automovilísticos según el día de la semana utilizando un diagrama de barras.

```python
accidentes_dia = df.groupby(df['weekday']).count().date
accidentes_dia = accidentes_dia.sort_values()
```

```python
accidentes_dia.plot(kind='barh',figsize=(12,7), color='c', alpha=0.5)
plt.title('Cantidad de accidentes automovilísticos por día de la semana', fontsize=20)
plt.xlabel('Número de Accidentes',fontsize=16)
plt.ylabel('Día de la Semana',fontsize=16);
```

```python
fig, ax = plt.subplots(figsize=(12,7))
ax.barh(accidentes_dia.index, accidentes_dia, color='r')
plt.title('Cantidad de accidentes automovilísticos por día de la semana', fontsize=20)
plt.xlabel('Número de Accidentes',fontsize=16)
plt.ylabel('Día de la Semana',fontsize=16);
plt.show()
```

Como se muestra en la gráfica anterior, el número de accidentes automovilísticos disminuye el fin de semana.

### Cantidad de accidentes automovilísticos por día del año

```python
accidents = df.groupby(df['date'].dt.date).count().date
accidents.head()
```

```python
sundays = df.groupby(df[df['date'].dt.dayofweek==6].date.dt.date).count().date
sundays.head()
```

```python
friday = df.groupby(df[df['date'].dt.dayofweek==4].date.dt.date).count().date
friday.head()
```

```python
accidentes = df.groupby(df['date'].dt.date).count().date

accidentes.plot(figsize=(13,8), color='blue')

plt.scatter(sundays.index, sundays, color='green', label='domingo')

plt.scatter(friday.index, friday, color='red', label='viernes')

plt.title('Cantidad de accidentes automovilísticos por día del año', fontsize=20)
plt.xlabel('Fecha',fontsize=16)
plt.ylabel('Número de accidentes por día',fontsize=16);
plt.legend()
```

Como podemos observar, hay entre 10 y 50 accidentes por día y el número de accidentes el viernes es, por regla general, mucho más alto que el número de accidentes el domingo.

### Cantidad de accidentes automovilísticos por hora

```python
accidentes_hora = df.groupby(df['date'].dt.hour).count().date
accidentes_hora
```

```python
plt.figure(figsize=(12,7))
#accidentes_hora.plot(kind='bar',figsize=(12,7), color='orange', alpha=0.5)
plt.bar(accidentes_hora.index, accidentes_hora, color='c', alpha=0.5) 
plt.title('Cantidad de accidentes automovilísticos por hora', fontsize=20)
plt.xlabel('Horas',fontsize=16)
plt.ylabel('Número de Accidentes',fontsize=16);
```

Como podemos observar en el gráfico, el mayor número de accidentes ocurre en las primeras horas de la mañana 8–9 y entre 12 y 20.


### Usando un histograma:

```python
plt.figure(figsize=(10,6))
plt.hist(df['date'].dt.hour,  bins=10)
plt.title('Distribución de accidentes automovilísticos por hora', fontsize=20)
plt.xlabel('Horas',fontsize=16)
plt.ylabel('Frecuencia de Accidentes',fontsize=16);
```

### Cantidad de accidentes automovilísticos por día de la semana y hora.


```python
accidentes_hora_dia = df.groupby([df['hour'],df['date'].dt.dayofweek.rename('day')]).count().date
accidentes_hora_dia
```

```python

accidentes_hora_dia.unstack().plot(kind='barh', figsize=(16,26))

plt.legend(labels=[calendar.day_name[x] for x in range(0,7)],fontsize=16)
plt.title('Cantidad de accidentes automovilísticos por día de la semana y hora',fontsize=20)
plt.xlabel('Número de Accidentes',fontsize=16)
plt.ylabel('Horas',fontsize=16);
```

```python
accidentes_parte_dia = df.groupby([df['part_of_the_day'],df['date'].dt.dayofweek.rename('day')]).count().date
accidentes_parte_dia.unstack().plot(kind='bar', figsize=(10,8))
plt.legend(labels=[calendar.day_name[x] for x in range(0,7)],fontsize=10)
plt.title('Cantidad de accidentes automovilísticos por día de la semana y hora',fontsize=16)
plt.xlabel('Número de Accidentes',fontsize=16)
plt.ylabel('Horas',fontsize=16);
```

### Es mejor usar mapa de calor

```python
accidentes_hora_dia.to_frame()
accidentes_hora_dia = accidentes_hora_dia.reset_index()
accidentes_hora_dia.head()
```

```python
df_accidentes = accidentes_hora_dia.pivot_table(values='date',index='hour',columns='day')
df_accidentes
```

```python
import seaborn as sns
plt.figure(figsize=(11,7))
sns.heatmap(data=df_accidentes,  cmap = "inferno_r")
plt.xlabel([calendar.day_name[x] for x in range(0,7)])
plt.title('Cantidad de accidentes automovilísticos por día de la semana y hora',fontsize=16)
#plt.xlabel('Día de la Semana',fontsize=16)
plt.xlabel([calendar.day_name[x] for x in range(0,7)])
plt.ylabel('Horas',fontsize=16);
```

```python
accidentes_parte_dia.to_frame()
df_accidentes_parte_dia = accidentes_parte_dia.reset_index()
df_accidentes_parte_dia.head()
```

```python
df_accidentes_parte_dia = df_accidentes_parte_dia.pivot_table(values='date',index='part_of_the_day',columns='day')
df_accidentes_parte_dia
```

```python
import seaborn as sns
plt.figure(figsize=(11,7))
sns.heatmap(data=df_accidentes_parte_dia,  cmap = "inferno_r")
plt.xlabel([calendar.day_name[x] for x in range(0,7)])
plt.title('Cantidad de accidentes automovilísticos por día de la semana y hora',fontsize=16)
#plt.xlabel('Día de la Semana',fontsize=16)
plt.xlabel([calendar.day_name[x] for x in range(0,7)])
plt.ylabel('Horas',fontsize=16);
```

## Análisis tipo de accidentes

Los datos que estamos analizando contienen información relacionada con 
* la fecha del accidente
* el tipo de accidente,
* la ubicación del accidente. 

Con respecto al tipo de accidente, el marco de datos incluye información como el número de víctimas, el número de vehículos involucrados en el accidente y el tipo de lesiones (leves o graves). Como antes, podemos examinar la distribución de todas esas variables utilizando gráficos de barras.

### Cantidad de Vehículos Involucrados

```python
vehiculos_involucrados = df.vehicles_involved.value_counts()
vehiculos_involucrados.sort_values()
```

```python
vehiculos_involucrados[12]=0
#Está ordenado por cantidad de vehículos involucrados 
vehiculos_involucrados.sort_index(inplace=True)
plt.figure(figsize=(12,7))
plt.bar(vehiculos_involucrados.index, vehiculos_involucrados,  color='darkblue', alpha=0.5)
plt.title('Cantidad de vehículos involucrado por accidente',fontsize=20)
plt.xlabel('Vehículos involucrados',fontsize=16)
plt.ylabel('Número de Accidentes',fontsize=16);

for index in vehiculos_involucrados.index:
    plt.text(x=index,y=vehiculos_involucrados.loc[index],s=str(vehiculos_involucrados.loc[index]),horizontalalignment='center')
```

La gráfica anterior muestra la cantidad de accidentes en 2017 de acuerdo con la cantidad de vehículos involucrados. En la mayoría de los accidentes, estuvieron involucrados dos vehículos (7028 accidentes en 2017). Además, la policía registró accidentes automovilísticos en los que estuvieron involucrados hasta 14 vehículos; sin embargo, los accidentes automovilísticos con muchos vehículos no son comunes.


### Cantidad de lesiones por tipo


El dataframe incluye información sobre cuántas víctimas sufrieron lesiones leves y graves en cada accidente automovilístico. Podemos representar fácilmente el porcentaje de lesiones leves y graves utilizando un diagrama circular de la siguiente manera:


```python
lesiones = df[['mild_injuries','serious_injuries']].sum()
lesiones
```

```python

plt.figure(figsize=(7,7))
plt.pie(lesiones, colors=['yellow','red'], labels=None, autopct='%1.1f%%')
plt.legend(labels=['Lesiones Leves', 'Lesiones Serias'])
plt.title('Tipo de Lesiones', fontsize=16)
plt.ylabel('')
```

La trama muestra que solo el 2% de las lesiones son lesiones graves. Aunque la mayoría de las lesiones en accidentes automovilísticos fueron leves, sería interesante analizar en qué circunstancias (hora, fecha, lugar) las lesiones graves son más frecuentes.

### Tasa de lesiones tipo por día de la semana

Número de lesiones graves por día de la semana.

```python
accidentes_serios = df[df['serious_injuries']!=0].groupby(df['date'].dt.dayofweek).sum().serious_injuries
accidentes_serios
```

Tasa de lesiones graves por día de la semana.

```python
tasa_serios = accidentes_serios/accidentes_serios.sum()
tasa_serios
```

Cantidad de lesiones leves por día de la semana

```python
lesiones_leves = df[df['mild_injuries']!=0].groupby(df['date'].dt.dayofweek).sum().mild_injuries
lesiones_leves
```

Tasa de lesiones leves por día de la semana.

```python
tasa_leves = lesiones_leves/lesiones_leves.sum()
tasa_leves
```

```python
# Combine ambas series como un dataframe para trazarlas como una barra de lado a lado
tasas = pd.DataFrame({'Lesiones graves':tasa_serios,'Lesiones leves':tasa_leves})
tasas
```

```python

tasas.plot(kind='bar',figsize=(12,7),color=['red','blue'],alpha=0.5)
plt.title('Tasa de lesiones tipo por día de la semana',fontsize=20)
plt.xlabel('Día de la semana',fontsize=16)
plt.ylabel('Tasa',fontsize=16)
plt.xticks(np.arange(7),[calendar.day_name[x] for x in range(0,7)]);
```

La siguiente gráfica muestra el porcentaje de lesiones según el día de la semana. Las lesiones leves siguen un patrón esperado, ya que presentan tasas más altas durante los días de semana, cuando ocurren más accidentes. Sin embargo, las lesiones graves presentan tasas altas los fines de semana, aunque el número promedio de accidentes los fines de semana (1656) es menor que durante los días laborables (1025). Esto indica que los accidentes los fines de semana tienden a ser más graves que durante los días laborables.
