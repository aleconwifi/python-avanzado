### df.iterrows()

A veces es posible que desee recorrer / iterar sobre el marco de datos de Pandas y realizar alguna operación en cada fila. Pandas tiene al menos dos opciones para iterar sobre las filas de un marco de datos. Veamos ejemplos de cómo recorrer el marco de datos de Pandas. 

```python
import pandas as pd
```

```python
data = {'nombre': ['Ricardo', 'Luis', 'Manuel', 'Miguel'],
'lenguaje': [10, 9, 8, 8],
'quimica': [7, 6, 5, 5],
'fisica': [8, 9, 10, 8]}

df_materias = pd.DataFrame(data)
df_materias
```

```python

for index,row in df_materias.iterrows():
   print(index)
```

```python
for index,row in df_materias.iterrows():
   print(row['fisica'])
```

```python
vector = []
for index,row in df_materias.iterrows():
    vector.append((index, row['nombre'], row['fisica']))
vector
```

```python

for index,row in df_materias.iterrows():
    if (row == 8).any():
        print((index, row['nombre']))
```

### Usando un Loop Standard

```python
def congestion_sur():
    df['Congestion'] = 99999
    for row in range(0, len(df)):
        if (df.nivel.iloc[row] == 'ROJO') &  ((df.autonomia.iloc[row] == 'ANDALUCIA') | (df.autonomia.iloc[row] == 'MURCIA')):
            df['Congestion'].iloc[row] = 'Congestión en el sur de España' 
        else:
            df['Congestion'].iloc[row] = 'No hay congestión en el Sur de España'
```

```python
congestion_sur()
```

```python
%%timeit
congestion_sur()
```

### Usando un iterrows

```python
%%timeit
vector = []
for index, row in df.iterrows():
    condicion1 = row['nivel'] == 'ROJO'
    condicion2 = (row['autonomia'] == 'ANDALUCIA') | (row['autonomia'] == 'MURCIA')
    if (((condicion1) & (condicion2))):
        vector.append('Congestión en el sur de España')
    else:
        vector.append('No hay congestión en el Sur de España')
df['Congestion'] = vector
```

El código tardó milisegundos en ejecutarse, que es 100 veces más rápido que el bucle estándar. Sin embargo, muchas personas aconsejan no usarlo porque todavía hay opciones más rápidas.

Está mal usar el iterrowsmétodo en Pandas.Si estuviera en el equipo de desarrollo de Pandas, no dudaría en depreciarlo y luego eliminarlo de la existencia.Hay dos problemas con iterrows:

### Problema 1. Los bucles en los pandas son un pecado.

El punto de usar algo más sofisticado que el ingenuo almacenamiento de datos es la capacidad de realizar operaciones vectorizadas. Cuando se trabaja con millones o miles de millones de filas, la diferencia en el rendimiento es asombrosa. Muchos sistemas informáticos modernos, especialmente aquellos que utilizan GPU, están diseñados específicamente para aprovecharde la velocidad de las operaciones vectorizadas al computar cosas en paralelo, lo que no se puede hacer con procesos iterativos.

### Problema 2. Hay mejores formas de hacer básicamente lo mismo.

El segundo problema iterrows que incluso cuando encuentra el 0.001% del tiempo que está bien iterar o simplemente se siente especialmente perezoso y no está dispuesto a obtener una solución vectorizada, iterrows tiene un rendimiento peor que cualquiera de las otras formas de interactuar. 



### El método apply (): mucho más rápido


apply no es más rápido en sí mismo, pero tiene ventajas cuando se usa en combinación con DataFrames. Esto depende del contenido de la expresión apply.


```python
def sur_congestion(row):
    condicion1 = row['nivel'] == 'ROJO'
    condicion2 = (row['autonomia'] == 'ANDALUCIA') | (row['autonomia'] == 'MURCIA')
    if (condicion1) & (condicion2):
        resultado =  'Congestión en el sur de España'
    else:
        resultado = 'No hay Congestión en el Sur de España'
    return resultado
```

### Este código es incluso más rápido que los métodos anteriores

```python
%%timeit
df['Congestion'] = df.apply(lambda x: sur_congestion(x), axis=1)
```

### Vectorización de pandas: 9000 veces más rápido


Ahora podemos llegar a un nuevo tema. Utilizamos las ventajas de la vectorización para crear códigos realmente rápidos. El punto es evitar los bucles de nivel de Python como en los ejemplos anteriores y usar código C optimizado que usa la memoria mucho más eficiente. Solo necesitamos modificar ligeramente la función:

```python
        
def trafico_vectorizado():
    df['Congestion'] = 'No hay congesión en el Sur de España'
    df.loc[(df['nivel'] == 'ROJO') &  ((df['autonomia'] == 'ANDALUCIA') | (df['autonomia']  == 'MURCIA')) , 'Congestion' ] =  'Congestión en el sur de España'
```

```python
%%timeit
df['Congestion'] = trafico_vectorizado()
```

En este caso ni siquiera necesitamos un bucle. Todo lo que tenemos que hacer es ajustar el contenido de la función. Ahora podemos pasar directamente la serie Pandas a nuestra función, lo que causa la enorme ganancia de velocidad.

### Vectorización de pandas usando numpy: lo más rápido que hay


En el ejemplo anterior pasamos la serie Pandas a nuestra función. Al agregar .values, recibimos una matriz Numpy:


```python
df['nivel'].values
```

```python
        
def trafico_vectorizado_numpy():
    df['Congestion'] = 'No hay congestión en el Sur de España'
    df.loc[(df['nivel'].values == 'ROJO') &  ((df['autonomia'].values == 'ANDALUCIA') | (df['autonomia'].values  == 'MURCIA')) , 'Congestion' ] =  'Congestión en el sur de España'
```

```python
%%timeit
df['Congestion'] = trafico_vectorizado_numpy()
```

```python
df['Congestion'] = trafico_vectorizado_numpy()
```

* Si está seguro de que necesita usar un bucle, trate de usar apply.
* De lo contrario, la vectorización siempre es preferible ya que es mucho más rápida
