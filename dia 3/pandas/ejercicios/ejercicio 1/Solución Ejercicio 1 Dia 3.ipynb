### Paso 15. ¿Cuánto fue la ganancia total en el dataset? (Ganancia = Precio * Cantidad de Artículos Vendidos)

```python
ganancia = (df_ventas['quantity']* df_ventas['item_price']).sum()

print('La ganancia fue de: $' + str(np.round(ganancia,2)))
```

### Paso 16. ¿Cuántos pedidos se hicieron en el período? (Varios productos de un mismo pedido tienen el mismo order_id)

```python
orders = df_ventas.order_id.value_counts().count()
orders
print('Se hicieron {} pedidos'.format(orders))
```

### Paso 17. ¿Cuál es el monto promedio de ingresos por pedido?

```python
# Solución 1
df_ventas['ingresos'] = df_ventas['quantity'] * df_ventas['item_price']
df_agrupados_id = df_ventas.groupby(by=['order_id']).sum()
monto_promedio = df_agrupados_id.mean()['ingresos']
print('El monto promedio de ingresos es de $ {:.2f} por pedido'.format(monto_promedio))
```

```python
# Solución 2
df_ventas['ingresos'] = df_ventas['quantity'] * df_ventas['item_price']
df_agrupados_id = df_ventas.groupby(by=['order_id'])['ingresos'].sum()
monto_promedio = df_agrupados_id.mean()
print('El monto promedio de ingresos es de $ {:.2f} por pedido'.format(monto_promedio))
```

### Paso 21. ¿Cuántas veces se ordenó una Veggie Salad Bowl?
