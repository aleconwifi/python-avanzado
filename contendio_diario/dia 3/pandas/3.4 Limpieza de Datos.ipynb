```python
# Cargamos un dataset de incendios en la Selva del Amazonas
import pandas as pd

file_name = "https://raw.githubusercontent.com/rajeevratan84/datascienceforbusiness/master/amazon_fires.csv"
df = pd.read_csv(file_name, encoding = "ISO-8859-1")

df.head()
```

```python
# How many regions are in the dataset?

df['estado'].unique()
```

# Renombrando Columnas

```python
new_columns = {'ano' : 'año',
               'numero': 'número',
               'encontro': 'fecha'}

df.rename(columns = new_columns, inplace=True)
```

```python
df.tail()
```

```python
# ¿Cuántos distintos años existen?
df['año'].unique()
```

```python
# Exploremos nuestros tipos de fecha, debemos esperar que número sea un tipo de datos entero o flotante

df.info()
```

# Reorganizar columnas

```python

# Las columnas están numeradas de 0, de izquierda a derecha
# Pongamos la fecha primero, segundo el mes y tercer el año

nuevo_orden = [4,1,0,2,3,]
df = df[df.columns[nuevo_orden]]
df.head()
```

```python
#Otra forma de Ordenar 
df = df[['fecha', 'mes', 'año', 'estado', 'número']]
df.head()
```

```python
df.tail()
```


# Determinar si una columna contiene datos numéricos

```python
# No todos los datos son númericos
df['número'].str.isnumeric()
```

```python
df[df['número'].str.isdigit()]
```

```python

# Obtenemos el error anterior porque nuestro isdigit () devuelve Nan para valores en blanco o faltantes
# Para solucionar esto, necesitamos convertir nuestro tipo de datos de columna de objetos no nulos a un string

df['número'].astype(str).str.isdigit()
```



- Básicamente, `str.isdigit` solo devuelve True para las cadenas que contienen únicamente los dígitos 0-9.
- Por el contrario, `str.isnumeric` devuelve True si contiene caracteres numéricos.

# Manejo de datos faltantes

### Método Reemplazarlos con 0

```python
# Carguemos de nuevo el dataframe
file_name = "https://raw.githubusercontent.com/rajeevratan84/datascienceforbusiness/master/amazon_fires.csv"
df = pd.read_csv(file_name, encoding = "ISO-8859-1")


nuevos_nombres = {'ano' : 'año',
               'numero': 'número',
               'encontro': 'fecha'}


df.rename(columns = nuevos_nombres, inplace=True)
df['número'] = df['número'].str.strip(" Fires")

#creando una copia del dataframe
df_copy = df.copy()
df.head()
```

```python
# Visualización de la suma de valores faltantes en cada columna

df.isnull().sum()
```

```python

# Podemos eliminar fácilmente valores Nulos o NaN (no un número)
# Soltar filas con valores NaN

df = df.dropna() 
df = df.reset_index() #restablecer los índices de fila en caso de que se elimine alguna fila
df.head()
```

```python
# Vamos a ver y funcionó
df.isnull().sum()
```

### Bien, así que funcionó, ahora volvamos a cargar los datos y veamos algunos otros métodos para tratar con valores NaN o Null


```python
# Volvamos a cargar nuestro dataframe

file_name = "https://raw.githubusercontent.com/rajeevratan84/datascienceforbusiness/master/amazon_fires.csv"
df = pd.read_csv(file_name, encoding = "ISO-8859-1")
new_columns = {'ano' : 'año',
               'numero': 'número',
               'encontro': 'fecha'}
df.rename(columns = new_columns, inplace=True)
df['número'] = df['número'].str.strip(" Fires")
df_copy = df.copy()
df.head()
```

```python
# Crear un índice booleano para todos los valores nulos
df['número'].isnull()
```

```python
df[df['número'].isnull()].head(10)
```


# ¿Qué hacer con los datos faltantes?
* Eliminar usando .dropna(axis=0)
* Reemplazarlas con algún número arbitrario (por ejemplo, un promedio)
* Reemplácelos con Forward Fill (relleno posterior ) o Back Fill (relleno anterior)

```python
# Usnado fillna con ceros
df['número'].fillna(0).head()
```

```python
# Usado back filling
df['número'].fillna(method='ffill').head(70)
```

```python
# Ver el índice 444 para ver cómo cambia
# Tarea, cambie 444 usando ffill y backfill para ver cómo cambia
df.iloc[444]
```

```python
df.iloc[445]
```

```python
df['número'] = df['número'].fillna(method='bfill')
```

```python

# supongamos que los valores en blanco son incendios 0
# recuperemos nuestra copia de nuestros datos originales preprocesados

df = df_copy

# replace all missing or NaN values with 0
df['número'] = df['número'].fillna(0)
```

```python
# Vamos a ver si cambiamos nuestros Nans a 0s
df.iloc[444]
```


# Reemplazar texto en las columnas

```python
# Let's convert our Portuguese month names to English

mes_esp = {'Janeiro': 'Enero',
'Fevereiro': 'Febrero',
'Março': 'Marzo',
'Abril': 'Abril',
'Maio': 'Mayo',
'Junho': 'Junio',
'Julho': 'Julio',
'Agosto': 'Agosto',
'Setembro': 'Septiembre',
'Outubro': 'Octubre',
'Novembro': 'Noviembre',
'Dezembro': 'Diciembre'}

df["mes"] = df["mes"].map(mes_esp)
df.head()
```

# Eliminando Filas

Usando la función df.index

```python
# Vamos a borrar la primera fila
df = df.drop(df.index[0])
df = df.reset_index()
df.head()
```

```python
# Borrar múltiples filas

df = df.drop(df.index[[2,3]])
df.head()
```

```python
# Borrar un rango de filas

df = df.drop(df.index[1:4])
df.head()
```
