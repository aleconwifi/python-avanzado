```python
import pandas as pd

# Volvamos a cargar nuestro dataframe
file_name = "https://raw.githubusercontent.com/rajeevratan84/datascienceforbusiness/master/amazon_fires.csv"
df = pd.read_csv(file_name, encoding = "ISO-8859-1")
columnas_esp = {'ano' : 'año',
               'numero': 'número',
               'encontro': 'fecha'}
df.rename(columns = columnas_esp, inplace=True)
df['estado'] = df['estado'].str.title()
df_copy = df.copy()

df['número'] = df['número'].str.strip(" Fires")
df['número'] = df['número'].fillna(0)
df['número'] = df['número'].astype(float)

# Introuding a la función map
mes_esp = {'Janeiro': 'Enero',
'Fevereiro': 'Febrero',
'Março': 'Marzo',
'Abril': 'Abril',
'Maio': 'Mayo',
'Junho': 'Junio',
'Julho': 'Julio',
'Agosto': 'Agosto',
'Setembro': 'Septiembre',
'Outubro': 'Octubre',
'Novembro': 'Noviembre',
'Dezembro': 'Diciembre'}

df["mes"] = df["mes"].map(mes_esp)
df.head(15)
```


# Uso de la función Groupby

Esta es una de las herramientas de análisis de datos más útiles proporcionadas por Pandas. Nos permite agrupar segmentos comunes de datos y realizar una operación. Por ejemplo, podemos agrupar nuestros 'Estados' y encontrar la suma total de incendios por estado.

```python

# Primero agreguemos la fecha en función de cada estado
# Crearemos nuestro objeto groupby

df.groupby('estado')
```

```python

# ¿Cómo usamos nuestra objeto?

estado_agrupados = df.groupby('estado')
estado_agrupados.get_group('Rio')
```

```python

# Veamos los grupos que hicimos

# Es un diccionario donde cada clave corresponde a un nombre de estado
# Los valores son los rangos de índice para cada grupo

estado_agrupados.groups
```

```python
df.groupby(['estado']).groups.keys()
```

```python
# Podemos usar Tamaño para obtener un recuento de la cantidad de elementos en cada grupo
estado_agrupados.size()
```

```python
estado_agrupados.first()
```

# Ahora usemos algunas funciones de agregación
* mean()
* size()
* count()

```python
#estado_agrupados.sum()
df.groupby('estado').sum()
```


Tenga en cuenta que la celda superior izquierda llamada 'estado' ahora es el índice Anteriormente el índice estaba numerado numéricamente. Las columnas año y número son los medios agrupados

```python
estado_agrupados.mean()
```

# Pivot Table

Para las usuarios de Excel, puede estar pensando esto. Esta funcionalidad nos permite agrupar, ordenar, calcular datos y manejar datos de una forma muy similar a la que se hace con las hojas de cálculo.

```python
import numpy as np

df.pivot_table(values='número', index='estado', aggfunc=np.mean)
```

```python

# Otra característica proporcionada por la función Pivot
# es que podemos agregar "Todos" (ver última línea) que da la media de Todos los Estados

df.pivot_table(values='número', index='estado', aggfunc=np.mean, margins=True)
```

#  Agregando múltiples columnas usando Pivot Tables

```python
df.shape[0]
```

```python
# Antes de comenzar, creemos una columna de fecha falsa llamada "severidad"
# en esta columna usaremos una calificación de 1 a 5 para indicar la gravedad de cada incendio

import numpy as np

df['severidad'] = np.random.randint(1, 5, df.shape[0])
df.head(10)
```

## Formato para agregar múltiples columnas de datos


**DataFrame.pivot_table([column1_values, column2_values], column_index)**

```python
# Anteriormente lo hicimos asi:
# df.pivot_table(values='número', index='estado', aggfunc=np.mean)

df.pivot_table(['número', 'severidad'], 'estado')
```

```python

df.pivot_table(values=['número', 'severidad'], index='estado', aggfunc=np.mean, margins=True)
```

```python
# Aggregating with multiple functions on multiple columns

df.pivot_table(['número', 'severidad'], 'estado', aggfunc=[np.mean, np.min, np.max], margins=True)
```

# Aprendiendo a usar GroupBy Agg

### La sintaxis es la siguiente:

**DataFrame.groupby(index)[column1].agg(function)**

```python
# De manera similar, podemos usar la función 'agg' con groupbys
df.groupby('estado')['número'].agg(np.mean)
```

```python
## Aplicandolo a múltiples columnas
df.groupby('estado')['número', 'severidad'].agg(np.mean)
```

## Aplicando múltiples funciones de agregación


```python
df.groupby('estado')['número', 'severidad'].agg([np.mean, np.max])
```

```python
df.groupby('estado').agg(
    {
         'número':sum, 
         'severidad': "mean",
         'fecha': 'first'  
    }
)
```

```python
df.groupby(['estado', 'mes']).agg(
    {
         'número':[min, max, sum], 
         'severidad': "mean",
         'fecha': 'first'  
    }
)
```

# Usando funciones de Agregación modificadas

```python
def max_minus_mean(grouped_data):
    return(grouped_data.max() - grouped_data.mean())

df.groupby('estado')['número'].agg(max_minus_mean)
```

# Convertir de Series a Dataframe

```python
# Converting Serie a un DataFrame
#Forma 1
pd.DataFrame(df.groupby('estado')['número'].agg(max_minus_mean)).reset_index()
```

```python
#Forma 2
df.groupby('estado')[['número']].agg(max_minus_mean).reset_index()
```

```python
df.groupby('estado')['número'].agg(max_minus_mean).to_frame().reset_index()
```

## Nombrado a Columnas de Agregación

```python
df.groupby(['estado', 'mes']).agg(
    maximo_por_mes=('número', max),
    media_severidad=('severidad', np.mean),

)
```
