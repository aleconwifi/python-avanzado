### Colores a Filas del Dataframe

```python
import pandas as pd
import numpy as np

np.random.seed(24)
df = pd.DataFrame({'A': np.linspace(1, 10, 10)})

df = pd.concat([df, pd.DataFrame(np.random.randn(10, 4), columns=list('BCDE'))],
               axis=1)
df.iloc[0, 2] = np.nan
df
```

```python

def valor_columnaB(s):
    if s.B > 1.0:
        return ['background-color: yellow']*5
    else:
        return ['background-color: white']*5


df.style.apply(valor_columnaB, axis=1)
```

```python
def rojo_color_negativo(val):
    color = 'red' if val < 0 else 'black'
    return 'color: %s' % color

df.style.applymap(rojo_color_negativo)
```

```python
def color_rojo_maximo(s):
    is_max = s == s.max()
    return ['background-color: red' if v else '' for v in is_max]

df.style.apply(color_rojo_maximo)
```

```python
import seaborn as sns
cm = sns.light_palette("red", as_cmap=True)
df.style.background_gradient(cmap='viridis')
```

```python
df.style.bar(subset=['A', 'B', 'C', 'D', 'E'], color='#d65f5f')
```

### ipywidgets


Ahora veremos una introducción a los widgets de IPython. Estas son herramientas que nos permiten crear interactividad en nuestros cuadernos a menudo con una sola línea de código. Estos widgets son muy útiles para la exploración y análisis de datos, por ejemplo, seleccionando ciertos datos o actualizando gráficos. En efecto, los widgets le permiten convertir Jupyter Notebooks en un tablero interactivo en lugar de un documento estático.

```python
import ipywidgets as widgets
from ipywidgets import interact, interact_manual
```

```python
df = pd.read_csv('titanic.csv')
df.head()
```

```python
df.loc[df['age'] > 10]
```

```python
@interact
def mostrar_edad(column='age', x=30):
    return df.loc[df[column] > x]
```

```python
@interact
def show_articles_more_than(column=['age', 'pclass', 'fare', 'sibsp'], 
                            x=(1, 200, 1)):
    return df.loc[df[column] > x]
```
